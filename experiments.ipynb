```python
from transformers import CLIPTokenizer
import torch

# Initialize tokenizer
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")

# 1. Print the special token attributes
print("Special token map:", tokenizer.special_tokens_map)
print("Padding token ID:", tokenizer.pad_token_id)
print("EOS token ID:", tokenizer.eos_token_id)
print("BOS token ID:", tokenizer.bos_token_id)

# 2. Tokenize a short sentence without padding
caption = "a cat on a mat"
encoded = tokenizer(caption, return_tensors="pt")
print("\nTokenization without padding:")
print("Input IDs:", encoded.input_ids)
print("Token names:", tokenizer.convert_ids_to_tokens(encoded.input_ids[0]))
print("Attention mask:", encoded.attention_mask)

# 3. Tokenize with padding to see the difference
captions = ["a cat on a mat", "dog"]
padded = tokenizer(captions, padding="max_length", max_length=10, truncation=True, return_tensors="pt")
print("\nTokenization with padding:")
print("Input IDs:", padded.input_ids)
print("Tokens (first sentence):", tokenizer.convert_ids_to_tokens(padded.input_ids[0]))
print("Tokens (second sentence):", tokenizer.convert_ids_to_tokens(padded.input_ids[1]))
print("Attention mask (first):", padded.attention_mask[0])
print("Attention mask (second):", padded.attention_mask[1])

# 4. Explicitly add EOS and see how it appears in attention mask
caption_with_eos = "a cat on a mat"
encoded_with_eos = tokenizer(caption_with_eos, padding="max_length", max_length=10, return_tensors="pt")
# Manually add EOS token for demonstration
input_ids = encoded_with_eos.input_ids[0]
attention_mask = encoded_with_eos.attention_mask[0]

# Print token by token with attention mask values
print("\nToken-by-token inspection with attention mask:")
for i, token_id in enumerate(input_ids):
    token = tokenizer.convert_ids_to_tokens([token_id])[0]
    mask_value = attention_mask[i].item()
    print(f"Position {i}: Token ID {token_id} = '{token}', Attention mask: {mask_value}")
```
